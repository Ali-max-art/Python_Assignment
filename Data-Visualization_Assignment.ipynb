```python
#2). Generating a Heatmap to Visualize Sales Variation Across Different Months and Days
np.random.seed(20)
data_sales = {
    'Month': np.random.choice(['Jan', 'Feb', 'March', 'April', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data_sales)
print(df.head())
sales_pivot = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='sum')
print(sales_pivot)
import plotly.graph_objects as go
fig = go.Figure(data=go.Heatmap(
    z=sales_pivot.values,
    x=sales_pivot.columns,
    y=sales_pivot.index,
    colorscale='Viridis'))
fig.update_layout(
    title='Sales Variation Across Different Months and Days',
    xaxis_nticks=36,
    xaxis_title='Month',
    yaxis_title='Day'
)
fig.show()
```

### BOKEH ASSIGNMENT:

```python
#Q1). Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

#A1). Creating a Bokeh plot

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()

x = np.linspace(0,10,100)
y = np.sin(x)

p = figure(title= 'Sin Wave', x_axis_label = 'x', y_axis_label = 'sin(x)', width = 600, height = 400 )
p.line(x,y,legend_label = 'sin(x)', line_width = 2, color = 'blue')
show(p)
```

```python
#Q2). Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

#A2). Creating a Bokeh Scatter plot
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()
np.random.seed(42)
x = np.random.rand(100) * 100
y = np.random.rand(100) * 100
sizes = np.random.randint(10, 30, 100)
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], 100)
df = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'color': colors})
p = figure(title='Random Scatter Plot with Different Sizes and Colors', 
           x_axis_label='X values', y_axis_label='Y values', 
           width=600, height=400)
p.scatter('x', 'y', size='sizes', color='color', source=df, legend_field='color', fill_alpha=0.6)
show(p)
```

```python
#Q3).  Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
# fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
# counts = [20, 25, 30, 35]



#A3). Generating a Bokeh bar chart representing the counts of different fruits using the following dataset.

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource
from bokeh.palettes import Spectral6
output_notebook()
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
source = ColumnDataSource(data = dict(fruits = fruits, counts = counts))
p.vbar(x = 'fruits', top = 'counts', width = 0.5, source = source, legend_field="fruits", line_color='white', 
       fill_color=factor_cmap('fruits', palette=Spectral6, factors=fruits))
p.legend.title = 'Fruits'
p.legend.orientation = "horizontal"
p.legend.location = "top_center"
show(p)
```

```python
#Q4). Create a Bokeh histogram to visualize the distribution of the given data
# data_hist = np.random.randn(1000)
# hist, edges = np.histogram(data_hist, bins = 30)

#A3). Creating a Bokeh histogram
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
output_notebook()
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins = 30)
source = ColumnDataSource(data = dict( top = hist, left = edges[:-1], right = edges[1:]))
p = figure(title = 'Histogram of Randomly Generated Data', x_axis_label = 'Data Values', y_axis_label = 'Frequency', tools = '')
p.quad(bottom=0, top='top', left='left', right='right', source=source,
       fill_color='navy', line_color='white', alpha=0.7)
show(p)
                          

```

```python
#Q5). Create a Bokeh heatmap using the provided dataset.
# data_heatmap = np.random.rand(10, 10)
# x = np.linspace(0. 1, 10)
# y = np.linspace(0, 1, 10)
# xx, yy = np.meshgrid(x ,y)

#A5). Creating a Bokeh heatmap using the provided dataset.
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.layouts import layout
from bokeh.models import ColumnDataSource
output_notebook()
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
x_flat = xx.flatten()
y_flat = yy.flatten()
values = data_heatmap.flatten()
source = ColumnDataSource(data = dict(x= x_flat, y = y_flat, values = values ))
color_mapper = LinearColorMapper(palette = 'Viridis256', low = min(values), high = max(values))
p = figure(title="Heatmap of Random Data", x_axis_label='X', y_axis_label='Y', tools="", toolbar_location=None, x_range=(0, 1), y_range=(0, 1))
p.rect(x='x', y='y', width=1/10, height=1/10, source=source, fill_color=linear_cmap('values', 'Viridis256', min(values), max(values)), line_color=None)
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0), title='Value')
p.add_layout(color_bar, 'right')
show(p)
```
